PROBLEM STATEMENT: We need to read the image in RBG format and then convert it to a grayscale image. This will turn an image into a classic black and white photo. Then the next thing to do is invert the grayscale image also called negative image, this will be our inverted grayscale image. Inversion can be used to enhance details. Then we can finally create the pencil sketch by mixing the grayscale image with the inverted blurry image. This can be done by dividing the grayscale image by the inverted blurry image. Since images are just arrays, we can easily do this programmatically using the divide function from the cv2 library in Python.

## Importing Libraries

```python
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread("Manas_Dutta.jpg")
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

# Setting Figure Size
plt.figure(figsize=(12,5))
plt.imshow(img)

plt.title("Manas Dutta's Pic")
plt.show()
```

## Converting Into Gray Scale

```python
gray_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(12,5))

plt.imshow(gray_img,cmap='gray')
plt.title("Gray Image of Manas Dutta")
plt.show()
```

```python
## Image Inverting
```

```python
invert_img=cv2.bitwise_not(gray_img)
plt.figure(figsize=(12,5))

plt.imshow(invert_img,cmap='gray')
plt.title("Inverted Image of Manas Dutta")
plt.show()
```

```python
## Converting into Gaussian Blur
```

```python
blur_img=cv2.GaussianBlur(invert_img,(21,21),sigmaX=0,sigmaY=0)
plt.figure(figsize=(12,5))

plt.imshow(blur_img,cmap='gray')
plt.title("Gaussian Blur Image of Manas Dutta")
plt.show()
```

## Converting Into Pencil Image

```python
pencil_img=cv2.divide(gray_img,255-blur_img,scale=255)
plt.figure(figsize=(12,5))

plt.imshow(pencil_img,cmap='gray')
plt.title("Final Pencil Sketch of Manas Dutta")
plt.show()
```
